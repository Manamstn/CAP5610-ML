```python
#MANA MOSTAANI
# HW2_Problem2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#read the data frame
df = pd.read_csv("HW2_nonlinear_data.csv")

#cubic non-linear regression
def nonlinear_regression(x,y):

    a = np.random.randn()
    b = np.random.randn()
    c = np.random.randn()
    d = np.random.randn()

    alfa = 0.000001

    # predict = wx + b
    for i in range(10000):
        predict = a*np.power(x,3) + b*np.power(x,2) + c*x + d

        #MSE
        loss_func = np.mean((y - predict)**2)

        #Gradient Descent
        a = a - (alfa* np.mean((predict - y)*np.power(x,3)))
        b = b - (alfa* np.mean((predict - y)*np.power(x,2)))
        c = c - (alfa* np.mean((predict - y)*x))
        d = d - (alfa* np.mean(predict - y))

    return a,b,c,d
```

```python
a,b,c,d = nonlinear_regression(df["X"], df["Y"])
```

```python
#ploting the data
plt.scatter(df["X"],df["Y"])
#ploting the cubic non-linear regression
plt.scatter(df["X"],a*np.power(df["X"],3) + b*np.power(df["X"],2) + c*df["X"] + d, color="red")
```
